### Energy and RMSE

The energy (Wikipedia; FMP, p. 66) of a signal corresponds to the total magntiude of the signal. For audio signals, that roughly corresponds to how loud the signal is. The energy in a signal is defined as

$$\sum_n|x(n)|^2$$
 
The root-mean-square energy (RMSE) in a signal is defined as

$$\sqrt{\frac{1}{N}\sum_n|x(n)|^2}$$

### Strip
removes leading silence from a signal

```python
def strip(x,frame_length,hop_length):
    rmse = librosa.feature.rms(x,frame_length=frame_length,
                               hop_length=hop_length,center=True)
    thresh,frame_index = .01,0
    while rmse[0][frame_index]<thresh:
        frame_index += 1
    
    start_sample_index = librosa.frames_to_samples(frame_index,
                                                  hop_length=hop_length)
    return x[start_sample_index:]
```

```python
ystrip = strip(y,win_length,hop_length)
print(ystrip.shape)
```

### Mel-spectrogram

The human perception of sound intensity is logarithmic in nature. Therefore, like the STFT-based spectrogram, we are often interested in the log amplitude:

```python
S = librosa.feature.melspectrogram(y,sr=sr,n_fft=n_fft,
                                   hop_length=hop_length)
logS = librosa.power_to_db(abs(S))
librosa.display.specshow(logS, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel')
```

### Constant-Q transform
Unlike the Fourier transform, but similar to the mel scale, the constant-Q transform uses a logarithmically spaced frequency axis.

```python
fmin = librosa.midi_to_hz(36)
C = librosa.cqt(y,sr=sr,fmin=fmin,n_bins=72)
logC = librosa.amplitude_to_db(abs(C))
```

```python
librosa.display.specshow(logC, sr=sr, x_axis='time', y_axis='cqt_note', fmin=fmin, cmap='coolwarm')
plt.colorbar(format='%+2.0f dB')
```

### Chroma
A chroma vector (Wikipedia) (FMP, p. 123) is a typically a 12-element feature vector indicating how much energy of each pitch class, {C, C#, D, D#, E, ..., B}, is present in the signal.

#### Chroma STFT

```python
chromagram = librosa.feature.chroma_stft(y,sr=sr,hop_length=hop_length)
librosa.display.specshow(chromagram, x_axis='time', y_axis='chroma', hop_length=hop_length, cmap='coolwarm')
```

### Chroma energy normalized statistics (CENS) 
The main idea of CENS features is that taking statistics over large windows smooths local deviations in tempo, articulation, and musical ornaments such as trills and arpeggiated chords. CENS are best used for tasks such as audio matching and similarity.

```python
chromagram = librosa.feature.chroma_cens(y, sr=sr, hop_length=512)
librosa.display.specshow(chromagram, x_axis='time', y_axis='chroma', hop_length=hop_length, cmap='coolwarm')
```
