```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class mcEstimatePi:
    def __init__(self,N,r):
        self.N = N
        self.r = r
    def sampling(self):
        xs=np.random.uniform(-self.r,self.r,self.N)
        ys=np.random.uniform(-self.r,self.r,self.N)
        rs=xs**2+ys**2
        inside = (rs<=self.r**2)
        samples = 4*(self.r**2)*inside
        outside = ~inside
        
        Ihat = np.mean(samples)
        piHat = Ihat/(self.r**2)
        se = np.sqrt(np.var(samples)/self.N)
        return xs,ys,inside,outside,piHat,se,Ihat
```

```python
demo = mcEstimatePi(5000,2)
xs,ys,inside,outside,pi,std,Isum = demo.sampling()
```

```python
plt.scatter(xs[inside],ys[inside])
plt.scatter(xs[outside],ys[outside])
plt.axis([-2.03,2.03,-2.03,2.03])
plt.axes().set_aspect('equal')
```

```python
print('Estimated pi: '+str(pi))
print('Standard Deviation: '+str(std))
print('Estimated Integral: '+str(Isum))
```
