```python
import numpy as np 
import matplotlib.pyplot as plt
import math
import scipy.integrate as inte
```

```python
def trajRamp(t):
    return t
def trajSin(t):
    return np.sin(t)
```

```python
def trajStep(t):
    if isinstance(t,(int,float)):
        x = -1 if t > 5 else 1
    else:
        x = np.ones(t.shape)
        x = [-1 if t[i] > 5 else 1 for i in range(t.size)]
    return x
```

```python
def controller(params,t,x,xd):
    kp,kv = -3000,-100
    errT = x-params['traj'](t)
    u = kv*xd**3 + kp*errT
    return u
```

```python
def doubleIntegrator(params,t,X):
    u = controller(params,t,X[0],X[1])
    u = np.clip(u,-5,5)
    return np.array([X[1],u])
```

```python
def simAndPlot(trajToUse):
    params = {}
    params['traj'] = trajToUse
    t0,t1 = 0,10; X0 = np.array([0,0])
    t = np.linspace(t0,t1,100)
    X = np.zeros((len(t),len(X0)))
    X[0,:] = X0
    r = inte.ode(lambda t,X:doubleIntegrator(params,t,X)).set_integrator("dopri5") 
    r.set_initial_value(X0, t0)   # initial values
    for i in range(1, t.size):
       X[i, :] = r.integrate(t[i]) # get one more value, add it to the array
       if not r.successful():
           raise RuntimeError("Could not integrate")
    
    plt.plot(t,params['traj'](t),'r--')
    plt.plot(t,X[:,0])
    
    return t,X
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
t,X = simAndPlot(lambda t:trajRamp(t))
plt.subplot(2,2,2)
t,X = simAndPlot(lambda t:trajSin(t))
plt.subplot(2,2,3)
t,X = simAndPlot(lambda t:trajStep(t))
```
