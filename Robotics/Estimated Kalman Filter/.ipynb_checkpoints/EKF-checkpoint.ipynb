```python
import numpy as np
import matplotlib.pyplot as plt
import math
from numpy import matmul as mm
import scipy.io as sio
from scipy.linalg import cho_factor,cho_solve
mat_contents = sio.loadmat('data.mat')
trial0 = mat_contents['trial0']
trial1 = mat_contents['trial1']
```

```python
trial0.shape
```

```python
def cosd(theta):
    return np.cos(theta*np.pi/180)

def sind(theta):
    return np.sin(theta*np.pi/180)

cosd(180)
```

```python
def EKF(data):
    accel_sens, gyro_sens, g = 16384,131,1
    ayz = data[:,1:3]/accel_sens
    gx = data[:,-1].reshape(-1,1)/gyro_sens
    t = data[:,0]/1000
    z = np.hstack((ayz,gx))
    
    xhat = np.zeros((2,len(t)))
    P = np.zeros((2,2,len(t)))
    Q = np.array([[.1,0],[0,.1]])
    R = np.array([[[.009,0,0],[0,.009,0],[0,0,.0076]]])
    xhat[:,0] = np.array([2,0])
    P[:,:,0] = 5*np.ones((2,2))

    for i in range(1,len(t)):
        dt = t[i]-t[i-1]
        A = np.array([[1,dt],[0,1]])
        xkp = mm(A,xhat[:,i-1])
        Pkp = mm(mm(A,P[:,:,i-1]),A.T)+Q
        H = np.array([[g*cosd(xhat[0,i])*np.pi/180,0],
                      [-g*sind(xhat[0,i])*np.pi/180,0],
                      [0,1]])
        h = np.array([[g*sind(xkp[0])],
                      [g*cosd(xkp[0])],
                      [xkp[1]]])
        c,low = cho_factor((mm(mm(H,Pkp),H.T)+R).squeeze().T)
        K = cho_solve((c,low),mm(Pkp,H.T).T).T
        xhat[:,i]=(xkp.reshape(-1,1) + mm(K,z[i,:].reshape(-1,1)-h)).squeeze()
        P[:,:,i] = mm((np.eye(2)-mm(K,H)),Pkp)    
        
    return xhat,t,gx,ayz
```

```python
xhat,t,gx,ayz = EKF(trial0)
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.plot(t,xhat[0,:])

plt.subplot(2,2,2)
plt.plot(t,xhat[1,:])
plt.plot(t,gx)

accelPred = np.zeros((2,len(t)))
for i in range(1,len(t)):
    accelPred[:,i]=np.array([sind(xhat[0,i]),cosd(xhat[0,i])])
plt.subplot(2,2,3)
plt.plot(t,accelPred[0,:])
plt.plot(t,ayz[:,0])
plt.subplot(2,2,4)
plt.plot(t,accelPred[1,:])
plt.plot(t,ayz[:,1])
```

```python
xhat,t,gx,ayz = EKF(trial1)
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.plot(t,xhat[0,:])

plt.subplot(2,2,2)
plt.plot(t,xhat[1,:])
plt.plot(t,gx)

accelPred = np.zeros((2,len(t)))
for i in range(1,len(t)):
    accelPred[:,i]=np.array([sind(xhat[0,i]),cosd(xhat[0,i])])
plt.subplot(2,2,3)
plt.plot(t,accelPred[0,:])
plt.plot(t,ayz[:,0])
plt.subplot(2,2,4)
plt.plot(t,accelPred[1,:])
plt.plot(t,ayz[:,1])
```
