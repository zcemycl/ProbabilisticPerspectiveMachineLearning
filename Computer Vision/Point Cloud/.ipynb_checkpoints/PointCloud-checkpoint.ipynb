### Essential Matrix Estimation
Given $\mathrm{E}$ essential matrix, $\mathrm{K}$ camera intrinsic parameter matrix, $\mathrm{F}$ fundamental matrix, 
$$\mathrm{E=K^\top F K}$$
Due to noise in the intrinsic parameters, the singular values of the essential matrix are not necessarily $(1,1,0)$. The essential matrix can be corrected by reconstructing it with $(1,1,0)$ singular values, i.e. $\mathrm{E}=\mathrm{U}\begin{pmatrix}1&0&0\\0&1&0\\0&0&0\end{pmatrix}\mathrm{V^\top}$ and normalize the essential matrix such that $\|\mathrm{E}\|=1$.

```python
def EssentialMatrixFromFundamentalMatrix(F,K):
    E_ = np.matmul(np.matmul(K.T,F),K)
    u,s,vh = np.linalg.svd(E_)
    d = np.diag(np.array([1,1,0]))
    E = np.matmul(np.matmul(u,d),vh.T)
    return E
```

### Fundamental Matrix Estimation
Given $N\geq 8$ correspondences between two images, $\mathbf{x_1}\Leftrightarrow\mathbf{x_2}$, implement the following function that linearly estimates a fundamental matrix, $\mathbf{F}$, such that $\mathbf{x_2^\top Fx_1}=0$.

The fundamental matrix can be estimated by solving linear least squares ($\mathbf{Ax=0}$) by rearranging $\mathbf{F}$ as $\mathbf{x}$. Due to noise on correspondences, the estimated fundamental matrix can be rank 3. The last singular value of the estimated fundamental matrix must be set to zero to enforce the rank 2 constraint. Please also normalize the fundamental matrix such that $\|\mathbf{F}\|=1$.

### Linear Triangulation
The relative camera transform between image 1 ($\mathbf{R_1}=\mathbf{I_3}, \mathbf{C_1}=\mathbf{0_3}$) and image 2 ($\mathbf{R_2}$, $\mathbf{C_2}$) is provided. Given the relative transformation and point correspondences, $\mathbf{x_1}\Leftrightarrow\mathbf{x_2}$, triangulate 3D points using linear least squares,

This algorithm uses the two equations for perspective projection to solve for the 3D point that are optimal in a least squares sense.  https://www.uio.no/studier/emner/matnat/its/nedlagte-emner/UNIK4690/v16/forelesninger/lecture_7_2-triangulation.pdf

Each perspective camera model gives rises to two equations on the three entries of $\mathbf{X}_i$. 

Left camera:

\begin{align}
\mathbf{\tilde{u}}_i &= P\mathbf{\tilde{X}}_i \Longrightarrow \mathbf{\tilde{u}}_i\times P\mathbf{\tilde{X}}_i = 0 \\
\begin{pmatrix}  u_i \\ v_i \\ 1 \end{pmatrix} \times\begin{pmatrix}  \mathbf{p}^{1T} \\ \mathbf{p}^{2T} \\ \mathbf{p}^{3T} \end{pmatrix} \mathbf{\tilde{X}}_i &= 0 \Longrightarrow \begin{pmatrix} v_i\mathbf{p}^{3T}-\mathbf{p}^{2T} \\ \mathbf{p}^{1T}-u_i\mathbf{p}^{3T}  \\ u_i\mathbf{p}^{2T}-v_i\mathbf{p}^{1T}    \end{pmatrix} \mathbf{\tilde{X}}_i = 0 \\
\begin{pmatrix} v_i\mathbf{p}^{3T}-\mathbf{p}^{2T} \\ u_i\mathbf{p}^{3T}-\mathbf{p}^{1T}  \\    \end{pmatrix} \mathbf{\tilde{X}}_i &= 0
\end{align}

Similarly for right camera, $$\mathbf{\tilde{u'}}_i = P'\mathbf{\tilde{X}}_i \Longrightarrow \mathbf{\tilde{u'}}_i\times P'\mathbf{\tilde{X}}_i = 0 $$
$$\begin{pmatrix} v'_i\mathbf{p}^{'3T}-\mathbf{p}^{'2T} \\ u'_i\mathbf{p}^{'3T}-\mathbf{p}^{'1T}  \\    \end{pmatrix} \mathbf{\tilde{X}}_i = 0$$


Combining these equations, an over determined homogeneous system of linear equations that can be solved with SVD is obtained. 

\begin{align}
\begin{pmatrix} v_i\mathbf{p}^{3T}-\mathbf{p}^{2T} \\ u_i\mathbf{p}^{3T}-\mathbf{p}^{1T}  \\ v'_i\mathbf{p}^{'3T}-\mathbf{p}^{'2T} \\ u'_i\mathbf{p}^{'3T}-\mathbf{p}^{'1T}  \\    \end{pmatrix} \mathbf{\tilde{X}}_i &= 0 \\
A \mathbf{\tilde{X}}_i &= 0 
\end{align}

```python
def computeA(P1,P2,x1,x2):
    u=x1[0];v=x1[1];up=x2[0];vp=x2[1]
    A = np.array([[v*P1[2,:]-P1[1,:]],
                 [u*P1[2,:]-P1[0,:]],
                 [vp*P2[2,:]-P2[1,:]],
                 [up*P2[2,:]-P2[0,:]]])
    return A
    
def LinearTrianglulation(K,C1,R1,C2,R2,x1,x2):
    P1 = np.matmul(K,np.hstack((R1,C1)))
    P2 = np.matmul(K,np.hstack((R2,C2)))
    X = []
    for i in range(x1.shape[0]):
        A = computeA(P1,P2,x1[i,:],x2[i,:])
        _,_,vh = np.linalg.svd(A.squeeze())
        X.append(vh.T[:3,-1].T/vh[3,3])
        
    return np.array(X)
```

### Linear Camera Pose Estimation
Register the image 3 into the reconstructed coordinate system using 2D-3D correspondences, $\mathbf{x}\leftrightarrow\mathbf{X}$. The camera pose estimation uses linear least squares to compute the camera projection matrix, $\mathbf{P} \in \mathbb{R}^{3\times 4}$, and factors out camera intrinsic parameter: $\mathbf{K}^{-1}\mathbf{P} = \left[\begin{array}{cc}\mathbf{R}&\mathbf{t}\end{array}\right]$ where $\mathbf{t}$ is camera translation in the camera coordinate system, i.e., $\mathbf{C}=-\mathbf{R^\top}\mathbf{t}$. $\mathbf{R}$ must be cleaned up such that it is orthogonal matrix with the determinant 1 and $\mathbf{t}$ must be rescaled:

\begin{align}
& \mathbf{R} = \mathbf{UDV^\top} \\
& \begin{cases}
\mathbf{R_c} = \mathbf{UV^\top}, \mathbf{t_c} = \frac{\mathbf{t}}{\mathbf{D_{11}}} & \text{if} \det(\mathbf{UV^\top})=1 \\
\mathbf{R_c} = -\mathbf{UV^\top}, \mathbf{t_c} = -\frac{\mathbf{t}}{\mathbf{D_{11}}} & \text{if} \det(\mathbf{UV^\top})=-1
\end{cases}
\end{align}

where $\mathbf{D_{11}}$ is the first singular value of the $\mathbf{R}$. $\mathbf{R}_c$ and $\mathbf{t}_c$ are the cleaned-up rotation and translation for the camera.

```python
def reprojectionPts(X,K,R,C,R3,C3):
    x1p = np.matmul(K,X.T)
    x1p/=np.tile(x1p[2,:],(3,1))
    x2p = np.matmul(np.matmul(K,R),(X.T-np.tile(C,(1,X.shape[0]))))
    x2p/=np.tile(x2p[2,:],(3,1))
    x3p = np.matmul(np.matmul(K,R3),(X.T-np.tile(C3,(1,X.shape[0]))))
    x3p/=np.tile(x3p[2,:],(3,1))
    return x1p,x2p,x3p
```

### Nonlinear Triangulation
Given three camera poses and linearly triangulated points, $\mathbf{X}$, refine the locations of the 3D points that minimizes the reprojection error.

The reprojection error should be minimized, 

$$\min_{\mathbf{X}}\sum^3_{j=1}(\tilde{x}_i - \frac{u_i}{w_i})^2+(\tilde{y_i}-\frac{v_i}{w_i})^2$$
where 

$$
\begin{pmatrix}u_i\\v_i\\w_i\end{pmatrix} = \mathbf{K}\mathbf{R}_i(\mathbf{X}-\mathbf{C}_i)
$$

$\mathbf{C}_i$ and $\mathbf{R}_i$ are the $i^{th}$ camera optical center and orientation. The nonlinear optimization is solved by finding $\mathbf{\Delta X}$.

\begin{align}
\Delta\mathbf{X} &= \mathbf{(J^\top J)^{-1}\mathbf{J}^\top(b-f(X))} \\
\mathbf{b} &= \left[\begin{array}{cccccc}\tilde{x_1}&\tilde{y_1}&\tilde{x_2}&\tilde{y_2}&\tilde{x_3}&\tilde{y_3}\end{array}\right]^\top \\
\mathbf{f(X)} &= \left[\begin{array}{cccccc}\frac{u_1}{w_1}&\frac{v_1}{w_1}&\frac{u_2}{w_2}&\frac{v_2}{w_2}&\frac{u_3}{w_3}&\frac{v_3}{w_3}\end{array}\right]^\top
\end{align}

where $\mathbf{J}$ is the Jocabian for $\mathbf{X}$.

\begin{align}
\mathbf{J} &= \left[\begin{array}{ccc}\frac{\partial \mathbf{f_1}}{\partial \mathbf{X}}^\top&\frac{\partial \mathbf{f_2}}{\partial \mathbf{X}}^\top&\frac{\partial \mathbf{f_3}}{\partial \mathbf{X}}^\top\end{array}\right]^\top \\
\frac{\partial \mathbf{f}_i}{\partial \mathbf{X}} &= \frac{1}{w^2}\left[\begin{array}{c}w_i\partial_{\mathbf{X}}u_i-u_i\partial_{\mathbf{X}}w_i \\ w_i\partial_{\mathbf{X}}v_i-v_i\partial_{\mathbf{X}}w_i\end{array}\right] \\
\frac{\partial u_i}{\partial \mathbf{X}} &= \left[\begin{array}{ccc}fr_{11}+p_xr_{31}&fr_{12}+p_xr_{32}&fr_{13}+p_xr_{33}\end{array}\right] \\
\frac{\partial v_i}{\partial \mathbf{X}} &= \left[\begin{array}{ccc}fr_{21}+p_yr_{31}&fr_{22}+p_yr_{32}&fr_{23}+p_yr_{33}\end{array}\right] \\
\frac{\partial u_i}{\partial \mathbf{X}} &= \left[\begin{array}{ccc}r_{31}&r_{32}&r_{33}\end{array}\right] 
\end{align}

where $\mathbf{K} = \left[\begin{array}{ccc}f&0&p_x\\0&f&p_y\\0&0&1\end{array}\right]$ and $\mathbf{R}=\left[\begin{array}{ccc}r_{11}&r_{12}&r_{13}\\r_{21}&r_{22}&r_{23}\\r_{31}&r_{32}&r_{33}\end{array}\right]$.
Update via $\mathbf{X}_{i+1}=\mathbf{X}_i+\Delta \mathbf{X}$ to find the local minimum.

```python
def Jocabian_Triangulation(C,R,K,X):
    uvw = np.matmul(np.matmul(K,R),X-C)
    u = uvw[0]; v = uvw[1]; w = uvw[2]
    f = K[0,0]; px = K[0,2]; py = K[1,2]
    du=np.array([f*R[0,0]+px*R[2,0],
                f*R[0,1]+px*R[2,1],
                f*R[0,2]+px*R[2,2]])
    dv=np.array([f*R[1,0]+py*R[2,0],
                f*R[1,1]+py*R[2,1],
                f*R[1,2]+py*R[2,2]])
    dw = R[2,:]
    J = np.array([[w*du-u*dw],[w*dv-v*dw]])/w**2
    return J.squeeze()
def Single_Point_Nonlinear_Triangulation(K,C1,R1,C2,R2,C3,R3,x1,x2,x3,X):
    J1 = Jocabian_Triangulation(C1,R1,K,X)
    J2 = Jocabian_Triangulation(C2,R2,K,X)
    J3 = Jocabian_Triangulation(C3,R3,K,X)
    J = np.vstack((J1,J2,J3))
    
    uvw1 = np.matmul(np.matmul(K,R1),X.reshape(3,-1)-C1)
    uvw2 = np.matmul(np.matmul(K,R2),X.reshape(3,-1)-C2)
    uvw3 = np.matmul(np.matmul(K,R3),X.reshape(3,-1)-C3)
    u1 = uvw1[0]; v1 = uvw1[1]; w1 = uvw1[2] 
    u2 = uvw2[0]; v2 = uvw2[1]; w2 = uvw2[2] 
    u3 = uvw3[0]; v3 = uvw3[1]; w3 = uvw3[2] 
    
    b = np.hstack((x1,x2,x3)).reshape(6,-1)
    f=np.array([u1/w1,v1/w1,u2/w2,v2/w2,u3/w3,v3/w3]).reshape(6,-1)
    deltaX = np.matmul(np.linalg.inv(np.matmul(J.T,J)),J.T)
    deltaX = np.matmul(deltaX,b-f)
    X = X+deltaX.squeeze()
    return X
def nonLinearTriangulation(K,C1,R1,C2,R2,C3,R3,x1,x2,x3,X):
    for j in range(10):
        # TO DO: the convergence criteria
        Xlist = []
        for i in range(len(X)):
            X_refine=Single_Point_Nonlinear_Triangulation(K,C1,R1,
                    C2,R2,C3,R3,x1[i],x2[i],x3[i],X[i,:])
            Xlist.append(X_refine)
    return np.array(Xlist)
```
